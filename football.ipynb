```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
from seleniumbase import Driver
```

```python
def aggregate_win(list_of_results):
    sum = 0
    for result in list_of_results: # iterate over marks for all the subjects
        if result == 1:
            sum += 1 # count the number of subjects with more than 50 marks

    return sum

def aggregate_lost(list_of_results):
    sum = 0
    for result in list_of_results: # iterate over marks for all the subjects
        if result == 2:
            sum += 1 # count the number of subjects with more than 50 marks

    return sum

def aggregate_draw(list_of_results):
    sum = 0
    for result in list_of_results:
        if result == 0:
            sum += 1 

    return sum

def aggregate_total(list_of_results):
    sum = 0
    for result in list_of_results: 
        sum += 1

    return sum
```

```python
def if_home_win(row):
    score = row['Score']
    score_array_with_pen = score.split()
    if len(score_array_with_pen) > 2:
        score = score_array_with_pen[1]
    else:
        score = row['Score']
    score_array = score.split('–')

    #home win
    if int(score_array[0]) > int(score_array[1]):
        return 1
    elif int(score_array[0]) == int(score_array[1]):
        return 0
    else:
        return 2
```

```python
def crawl_html(url):
    driver = Driver(uc=True, headless=False)

    # open URL with a 6-second reconnect time to bypass the initial JS challenge
    driver.uc_open_with_reconnect(url, reconnect_time=6)
    page_html = driver.get_page_source()

    driver.close()

    return page_html

def fetch_football_data(url):
    page_html = crawl_html(url)

    matches = pd.read_html(page_html, match = "Scores & Fixtures")
    data = matches[0]

    clean_data = data[(data['Score'].notna()) & (data['Home'] != 'Home')]
    if clean_data.empty:
        raise ValueError("Empty dataframe after filter")
    
    clean_data['match_result'] = clean_data.apply(if_home_win, axis=1)

    match_result = clean_data[['Home', 'match_result']].copy()
    df_to_excel = clean_data.groupby('Home').agg(
        wins=('match_result', aggregate_win),
        loses=('match_result', aggregate_lost),
        draws=('match_result', aggregate_draw),
        total=('match_result', aggregate_total)
    )
    return df_to_excel.reset_index()
```
