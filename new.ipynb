```python
from selenium.webdriver.common.by import By
def accept_cookies(driver):
    cookies = driver.find_elements(By.ID, "onetrust-accept-btn-handler")
    if(len(cookies) > 0):
        cookies[0].click()
    else:
        print("No Cookies to Click")

def sort_string(string):
    string = ''.join(e for e in string if e.isalnum())
    string = string.lower()
    return string
```

```python
def aggregate_win(list_of_results):
    sum = 0
    for result in list_of_results: # iterate over marks for all the subjects
        if result == 1:
            sum += 1 # count the number of subjects with more than 50 marks

    return sum

def aggregate_lost(list_of_results):
    sum = 0
    for result in list_of_results: # iterate over marks for all the subjects
        if result == 2:
            sum += 1 # count the number of subjects with more than 50 marks

    return sum

def aggregate_draw(list_of_results):
    sum = 0
    for result in list_of_results:
        if result == 0:
            sum += 1 

    return sum

def aggregate_total(list_of_results):
    sum = 0
    for result in list_of_results: 
        sum += 1

    return sum
```

```python
from selenium.webdriver import ChromeOptions
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium import webdriver
import time


def driver_code(url):
    Capabilities = DesiredCapabilities.CHROME
    Capabilities["pageLoadStrategy"] = "normal"
    options = ChromeOptions()

    useragentarray = [
        "Mozilla/5.0 (Linux; Android 13) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.5672.76 Mobile Safari/537.36"
    ]

    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    # options.add_argument(f"--user-data-dir=./profile{driver_num}")

    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument("disable-infobars")
    options.add_argument("disable-blink-features=AutomationControlled")

    driver = webdriver.Chrome(
        options=options
    )
    driver.execute_script(
        "Object.defineProperty(navigator, 'webdriver', {get: () => undefined})"
    )

    driver.execute_cdp_cmd(
        "Network.setUserAgentOverride", {"userAgent": useragentarray[0]}
    )
    driver.set_window_size(390, 844)
    options.add_argument("--disable-popup-blocking")
    #     driver.execute_script(
    #         """setTimeout(() => window.location.href="https://www.bet365.com.au", 100)"""
    #     )
    driver.get(url)
    time.sleep(1)
    return driver
```

```python
import pandas as pd


def get_data(url):
    driver = driver_code(url)
    accept_cookies(driver)

    home_teams = [score.text.split('\n')[0] for score in driver.find_elements(By.CSS_SELECTOR,".event__homeParticipant")]
    home_scores = [int(score.text) for score in driver.find_elements(By.CSS_SELECTOR, ".event__score--home")]
    away_scores = [int(score.text) for score in driver.find_elements(By.CSS_SELECTOR, ".event__score--away")]
    results = [1 if home > away else 0 if home == away else 2 for home, away in zip(home_scores, away_scores)]
    
    my_columns = ['Home', 'match_result']
    new_dataframe = pd.DataFrame(columns = my_columns)
    new_dataframe['Home'] = home_teams
    new_dataframe['match_result'] = results

    df_to_excel = new_dataframe.groupby('Home').agg(
        wins=('match_result', aggregate_win),
        loses=('match_result', aggregate_lost),
        draws=('match_result', aggregate_draw),
        total=('match_result', aggregate_total)
    )

    driver.close()

    return df_to_excel.reset_index()
```

```python
from xlsxwriter.color import Color
pd.io.formats.excel.header_style = None

url_uae_pro = 'https://www.flashscore.com/football/united-arab-emirates/uae-league/results/'
url_uae_division = 'https://www.flashscore.com/football/united-arab-emirates/division-1/results/'

url_j1 = 'https://www.flashscore.com/football/japan/j1-league/results/'
url_j2 = 'https://www.flashscore.com/football/japan/j2-league/results/'
url_j3 = 'https://www.flashscore.com/football/japan/j3-league/results/'

url_k2 = 'https://www.flashscore.com/football/south-korea/k-league-2/results/'

url_thai_1 = 'https://www.flashscore.com/football/thailand/thai-league-1/results/'
url_thai_2 = 'https://www.flashscore.com/football/thailand/thai-league-2/results/'

links = [url_uae_pro, url_uae_division, url_j1,
          url_j2, url_j3, url_k2,
          url_thai_1, url_thai_2]

leagues = ['uae pro league', 'uae division 1', 'j1',
           'j2', 'j3', 'K league 2',
           'thai league 1', 'thai league 2']

col_names = [
    'TT',
    'Giải',
    'K thua',
    'Thua 1',
    'K thắng',
    'Thắng 1'
]

writer = pd.ExcelWriter("pandas_new.xlsx", engine="xlsxwriter")
start_row = 2
start_col = 4
max_length = 0
num_per_row = 0
idx = 1
for link, league in zip(links, leagues):
    df_empty = False
    while True:
        try:
            data = get_data(link)
        except ValueError as e:
            if str(e) == "Empty dataframe after filter":
                df_empty = True
                break
            
            print(e)
            continue
        else:
            break

    if df_empty:
        continue

    cols = [
        data.query('loses==0')['Home'],
        data.query('loses==1')['Home'],
        data.query('wins==0')['Home'],
        data.query('wins==1')['Home']
    ]

    for col, col_name in zip(cols, col_names):
        
        col_length = col.shape[0]
        if col_length > max_length:
            max_length = col_length

        col.to_excel(writer, startrow=start_row, startcol=start_col, index=False, header=False)

        start_col += 1

    idx_cell_format = workbook.add_format({
        'bold': True, 
        'italic': True, 
        'align': 'center', 
        'bg_color': Color("Green"), 
        'fg_color': Color("Red"),
        'pattern' : 4,
    })
    
    cell_format = workbook.add_format({
        'bold': True, 
        'italic': True, 
        'align': 'center'
    })
    if start_row == 2:
        start_col = 2
        workbook  = writer.book
        worksheet = writer.sheets['Sheet1']

        worksheet.write(start_row, start_col, idx, idx_cell_format)
        worksheet.write(start_row, start_col + 1, league, cell_format)
        for col_name in col_names:
            worksheet.write(start_row - 1, start_col, col_name, cell_format)
            start_col += 1
    else:
        start_col = 4
        workbook  = writer.book
        worksheet = writer.sheets['Sheet1']

        worksheet.write(start_row, start_col - 2, idx, idx_cell_format)
        worksheet.write(start_row, start_col - 1, league, cell_format)

    start_col = 4
    start_row += max_length
    max_length = 0
    idx += 1

writer.close()
```
